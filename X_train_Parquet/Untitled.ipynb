```python
import pandas as pd
import glob
import os

# setting the path for joining multiple files
files = os.path.join("X_train*.parquet")

# list of merged files returned
files = glob.glob(files)

print("Resultant parquet after joining all parquet files at a particular location...");

# joining files with concat and read_csv
X_train = pd.concat(map(pd.read_parquet, files), ignore_index=True)
```
